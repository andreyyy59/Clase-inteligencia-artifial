```python
df = pd.read_csv('https://raw.githubusercontent.com/adiacla/bigdata/refs/heads/master/vivienda_clasificacion.csv')
df
```

```python
import matplotlib.pyplot as plt

# Calculate value counts and store in a variable
room_counts = df['habitaciones'].value_counts()

# Create the bar plot using the index and values of room_counts
plt.figure(figsize=(8, 6))
plt.bar(room_counts.index, room_counts.values)

# Customize the plot
plt.xlabel("Number of Rooms")
plt.ylabel("Frequency")
plt.title("Bar Plot of Number of Rooms")
plt.xticks(room_counts.index)  # Ensure all unique room counts are shown on the x-axis

# Display the plot
plt.show()
```

```python
# prompt: borrar todos los registros que sean superior a 31 y dejarlo en el mismo df para area parqeo y mostrar la grafica

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/adiacla/bigdata/refs/heads/master/vivienda_clasificacion.csv')

# Filter out rows where 'Area_parqeo' is greater than 31
df = df[df['Area_parqeo'] <= 31]

# Create the boxplot
plt.figure(figsize=(8, 6))
df.boxplot(column='Area_parqeo')
plt.title('Boxplot of Area_parqueo (Values <= 31)')
plt.ylabel('Area_parqueo')
plt.show()
```

```python
# prompt: Crear un arreglo con la variable datos, con datos sinteticos usando numpy donde la columna 0 sea la edad y la columna 1 sea las notas de AI que van de 0 a 5, has que sean dos arreglos separados uno de edades y otro de notas con decimales que las personas sean de 18 a 25 aos concatenalos al final

import numpy as np

# Generate synthetic data for ages (18-25)
edades = np.random.randint(18, 26, size=20)

# Generate synthetic data for AI grades (0-5 with decimals)
notas_ai = np.random.uniform(0, 5, size=20).round(2)

# Concatenate the arrays (optional, you can keep them separate)
datos = np.column_stack((edades, notas_ai))

datos
```

```python
# prompt: calcular el promedio de la columna 1 de datos con numpy

# Assuming 'datos' is your NumPy array and you want the average of the first column (edades)
average_age = np.mean(datos[:, 0])
print(f"The average age is: {average_age}")
```
